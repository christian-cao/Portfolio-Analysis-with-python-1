```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
boeing = pd.read_csv('BA_data.csv')
ebay = pd.read_csv('EBAY_data.csv')

```

```python
data = { 'Date': boeing['Date'],
         'boeing': boeing['Close'],
         'ebay': ebay['Close']}
data = pd.DataFrame(data)
data
```

```python
returns = np.log(data.iloc[:,1:] / data.iloc[:,1:].shift(1) )
returns
```

```python
weights = np.array([0.5,0.5])
```

```python
pfolio_var = np.dot(weights.T, np.dot(returns.cov()*250, weights))
pfolio_var
```

```python
pfolio_vol = (np.dot(weights.T, np.dot(returns.cov()*250, weights)))**0.5
pfolio_vol
```

```python
print(str(round(pfolio_vol,5)*100)+'%')
```
