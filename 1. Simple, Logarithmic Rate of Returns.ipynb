```python
import pandas as pd
import numpy as np # Possibility of vectorized computation
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
netflix = pd.read_csv('NFLX_data.csv')
```

```python
netflix
```

<b> Simple Rate of Return </b> 

```python
netflix['simple_return']= (netflix['Close']/netflix['Close'].shift(1))-1 
# new column with the simple return for every day
```

```python
netflix['simple_return']
```

<b> Graph </b>

```python
netflix['simple_return'].plot(figsize=(10,5))
plt.show()
# although the most times in finances the atypical negative returns are more often than pisitives,  this is not the case.
# could be useful compare with the crisis dates
```

```python
avg_ret_dayly = netflix['simple_return'].mean()
# average of dayly returns
avg_ret_dayly
```

```python
avg_ret_anual = netflix['simple_return'].mean()*250 # 250 trading days per year
# average of anual returns
avg_ret_anual
```

```python
print(str(round(avg_ret_anual, 5)*100) + '%')
```

<b> Logarithmic Rate of Return </b> 
<p> More useful for a single stock  </p> 

```python
netflix['log_return']= np.log(netflix['Close']/netflix['Close'].shift(1)) 
# new column with the simple return for every day
```

```python
netflix['log_return']
```

<b> Graph </b>

```python
netflix['log_return'].plot(figsize=(10,5))
plt.show()
# although the most times in finances the atypical negative returns are more often than pisitives,  this is not the case.
# could be useful compare with the crisis dates
```

```python
avg_ret_dayly = netflix['log_return'].mean()
# average of dayly returns
avg_ret_dayly
```

```python
avg_ret_anual = netflix['log_return'].mean()*250 # 250 trading days per year
# average of anual returns
avg_ret_anual
```

```python
print(str(round(avg_ret_anual, 5)*100) + '%')
# itÂ´s smaller than simple returns
```
