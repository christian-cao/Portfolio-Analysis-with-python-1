```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
boeing = pd.read_csv('BA_data.csv')
nike = pd.read_csv('NKE_data.csv')

```

```python
data = { 'Date': boeing['Date'],
         'boeing': boeing['Close'],
         'nike': nike['Close']}
data = pd.DataFrame(data)
data
```

```python
(data.iloc[:,1:] / data.iloc[0][1:] * 100).plot(figsize = (15, 5))
plt.show()
```

```python
log_returns = np.log(data.iloc[:,1:] / data.iloc[:,1:].shift(1) )
log_returns
```

```python
log_returns.mean() * 250
```

```python
log_returns.cov() * 250
```

```python
# 1000 posible combinatios
```

```python
pfolio_returns = []
pfolio_volatilities = []

for x in range (1000):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    pfolio_returns.append(np.sum(weights*log_returns.mean())*250)
    pfolio_volatilities.append(np.sqrt(np.dot(weights.T , np.dot(log_returns.cov()*250, weights))))

pfolio_returns, pfolio_volatilities    
```

```python
pfolio_returns = []
pfolio_volatilities = []

for x in range (1000):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    pfolio_returns.append(np.sum(weights*log_returns.mean())*250)
    pfolio_volatilities.append(np.sqrt(np.dot(weights.T , np.dot(log_returns.cov()*250, weights))))

pfolio_returns = np.array(pfolio_returns)
pfolio_volatilities = np.array(pfolio_volatilities)

pfolio_returns, pfolio_volatilities  
```

```python
portfolios = pd.DataFrame({'Return': pfolio_returns, 'Volatility': pfolio_volatilities})
```

```python
portfolios.plot(x='Volatility', y='Return', kind='Scatter', figsize=(10,6));
plt.xlabel('Expected Volatility')
plt.ylabel('Expected Return')
```
