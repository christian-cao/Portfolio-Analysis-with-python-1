```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sp500 = pd.read_csv('sp500.csv')
nike = pd.read_csv('NKE_data.csv')

data = sp500.merge(nike, on='Date', how = 'inner')
data
```

```python
data = { 'Date': data['Date'],
        'sp500': data['Close_x'],
         'nike': data['Close_y']}
data = pd.DataFrame(data)
data
```

```python
returns = np.log(data.iloc[:,1:] / data.iloc[:,1:].shift(1) )
returns
```

```python
cov = returns.cov() * 250
cov
```

```python
cov_with_market = cov.iloc[0,1]
cov_with_market
```

```python
market_var = returns['sp500'].var() * 250 # sp500
market_var
```

```python
nike_beta = cov_with_market / market_var
nike_beta
```

```python
free_risk = 0.025
nike_er = free_risk + nike_beta *0.05
nike_er 

# Return expected buying 'boeing'
```

```python
sharpe = (nike_er - free_risk) / ( returns['nike'].std()*250 **0.5)
sharpe
```
