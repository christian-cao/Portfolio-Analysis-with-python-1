```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
netflix = pd.read_csv('NFLX_data.csv')
amazon= pd.read_csv('AMZN_data.csv')
nvidia = pd.read_csv('NVDA_data.csv')
fbook= pd.read_csv('FB_data.csv')
```

```python
data = { 'Date': netflix['Date'],
         'netflix': netflix['Close'],
         'amazon': amazon['Close'],
         'nvidia': nvidia['Close'],
         'fbook': fbook['Close'] }
data = pd.DataFrame(data)
data
```

<b> Return of the Portfolio</b>

```python
returns = (data.iloc[:,1:] / data.iloc[:,1:].shift(1) )-1
returns
# new Dataframe with the simple return for every day
```

```python
weights = np.array([0.25,0.25,0.25,0.25])
```

```python
np.dot(returns,weights) # matrix multiplication
```

```python
annual_returns = returns.mean() * 250
annual_returns
```

```python
np.dot(annual_returns, weights)
```

```python
portfolio1 = str(round(np.dot(annual_returns, weights), 5)*100) + '%'
print(portfolio1)
```

```python
weights2 = np.array([0.2,0.2,0.35,0.25])
portfolio2 = str(round(np.dot(annual_returns, weights2), 5)*100) + '%'
print(portfolio2)
```
