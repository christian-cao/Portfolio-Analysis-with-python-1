```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
adidas = pd.read_csv('ADI_data.csv')
nike = pd.read_csv('NKE_data.csv')

```

```python
data = { 'Date': adidas['Date'],
         'adidas': adidas['Close'],
         'nike': nike['Close']}
data = pd.DataFrame(data)
data
```

```python
returns = np.log(data.iloc[:,1:] / data.iloc[:,1:].shift(1) )
returns
```

<b> adidas </b>

```python
returns['adidas'].mean()
```

```python
returns['adidas'].mean()*250
```

```python
returns['adidas'].std() #standar deviation
```

```python
returns['adidas'].std()*250 **0.5
```

```python
returns['nike'].mean()*250
```
